```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

**3. Make the dilation operation and show the results on the image of the Structuring
Element shaped 5 x 5 square without using the OpenCV morphology library, thus
generating the results shows in the figure (Use image “j.png”)**

```python
img = cv.imread('/content/drive/MyDrive/PCVK_Fara/j.png')

kernel = np.ones((5,5), np.uint8)
dilation = cv.dilate(img, kernel, iterations = 1)

plt.subplot(121),plt.imshow(img),plt.title('Initial Image')
plt.subplot(122),plt.imshow(dilation),plt.title('Dilation Result')
```

**4. Make the erosion operation and show the results on the image of the Structuring
Element shaped 5 x 5 square without using the OpenCV morphology library, thus
generating the results shows in the figure (Use image “j.png”)**

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv.morphologyEx(img, cv.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initial Image')
plt.subplot(122), plt.imshow(erosion), plt.title('Erosion Result')
```
