```python
from google.colab import drive

# accessing my google drive
drive.mount('/content/drive')
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
path = '/content/drive/MyDrive/PCVK_Fara/ktp.jpeg'
img = cv.imread (path)
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

# Close Province
cv.rectangle(img,pt1=(170,15), pt2=(390,25), color=(179,0,255), thickness=15)
plt.imshow(img)

# Close District
cv.rectangle(img,pt1=(190,45), pt2=(370,55), color=(0,0,255), thickness=10)
plt.imshow(img)
```
